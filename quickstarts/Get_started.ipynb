```
import matplotlib.pyplot as plt
months=['jan','feb','mar','apr']
sales=[20,30,50,70]
profits=[10,30,40,70]
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(months,sales)
plt.title('month sales')
plt.xlabel('months')
plt.ylabel('sales')
plt.subplot(1,2,1)
plt.bar(months,profits)
plt.title('monthly profits')
plt.xlabel('months')
plt.ylabel('profits')
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd
dates = pd.date_range(start='2025-01-01',periods=6,freq='ME')
sales=[10,20,30,40,50,60]
plt.plot(dates,sales,marker='o',color='red')
plt.title('monthly sales(time-series data)')
plt.xlabel('dates')
plt.ylabel('sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
